```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['Solarize_Light2'])
from scipy import special
```

```python
def solExacta(x, u, t, Gamma):
    divisor = 2 * np.sqrt(Gamma * t)
    sol = 0.5 * (special.erfc((x - u * t)/ divisor) + 
                 np.exp(u * x) * np.exp(-Gamma) * special.erfc((x + u * t)/divisor))
    return sol
```

```python
def printValues(N, U, Gamma, Pe):
    print('N = {} ; Gamma = {} ; U = {}; Pe = {}'.format(N, Gamma, U, Pe))
```

```python
L = 2.5 # m
U = 1.0 # m/s
N = 100 # Número de nodos
dt = 0.002 # Paso de tiempo
Tmax = 1.0
Nmax = int(Tmax / dt)
x = np.linspace(0,L,N)
```

```python
Gamma = 0.1 # 0.1, 0.01, 0.001, 0.0001
Pe = U / Gamma
printValues(N, U, Gamma, Pe)

for k in range(100,Nmax+1, 100):
    ti = k * dt
    plt.plot(x, solExacta(x, U, ti, Gamma), label='$t$ = {:3.4f}'.format(ti))

plt.suptitle('Solución exacta', fontsize = 18, y = 1.0, va = 'top')
plt.title('Advección Difusión NO estacionaria: $\Gamma =${}, $Pe =${}'.format(Gamma, Pe), fontsize=12, color='blue')
plt.ylabel('$\psi$')
plt.xlabel('$x$')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(color='white')
#plt.savefig('Exacta_AdvDif_NoSta_Pe_{}.pdf'.format(int(Pe)), bbox_inches='tight')
```
