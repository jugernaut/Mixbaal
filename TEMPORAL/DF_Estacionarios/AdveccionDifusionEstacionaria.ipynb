```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['Solarize_Light2'])
```

```python
def solExacta(x, A, B, Gamma, U, L):
    return (np.exp(U * x / Gamma) - 1) / (np.exp(U * L / Gamma) - 1) * (B - A) + A
```

```python
def solExactaPe(x, A, B, Pe, L):
    return (np.exp(Pe * x) - 1) / (np.exp(Pe * L) - 1) * (B - A) + A
```

```python
def printValues(N, L, Gamma, U, Pe):
    print('N = {} ; L = {} ; Gamma = {} ; U = {}; Pe = {}'.format(N, L, Gamma, U, Pe))
```

```python
Gamma = 1.0
L = 1.0
A = 1.0
B = 0.0

rho = 1.0; u = 1.0
U = rho * u
Pe = U / Gamma

N = 100
```

```python
printValues(N, L, Gamma, U, Pe)
l = [0.25, 0.5, 0.75, 1.0, 1.5, 2.0, 3.0]
for li in l:
    x = np.linspace(0.0, li, N)
#    if i == 4:
#        plt.plot(x, solExactaPe(x, A, B, Pe, li), 'k-', lw=4, alpha=0.5, label='L ={:3.3f}'.format(li))
#    else:
    plt.plot(x, solExactaPe(x, A, B, Pe, li), '-', label='L ={:3.3f}'.format(li))

plt.suptitle('Solucin exacta', fontsize = 18, y = 1.0, va = 'top')
plt.title('Adveccin-Difusin estacionaria: $Pe$ = {}, $q = 0.0$'.format(Pe), fontsize=12, color='blue')
#plt.title('Solucin exacta con $L$ variable y $q$ = 0')
plt.ylabel('$\psi$')
plt.xlabel('$x$')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(color='white')
plt.savefig('Exacta_AdvDif_Lvariable.pdf', bbox_inches='tight')
```
