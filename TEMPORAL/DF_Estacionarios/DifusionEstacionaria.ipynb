```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['Solarize_Light2'])
```

```python
def solExacta(x, A, B, Gamma, L, q):
    return ((B-A)/L+(q/(2*Gamma))*(L-x))*x+A
```

```python
def printValues(N, L, Gamma, q):
    print('N = {} ; L = {} ; Gamma = {} ; q = {}'.format(N, L, Gamma, q))
```

```python
Gamma = 1.0
L = 1.0
A = 1.0
B = 0.0
q = 0

N = 100
```

```python
printValues(N, L, Gamma, q)
#plt.figure(figsize=(12,5))
l = [0.25, 0.5, 0.75, 1.0, 1.5, 2.0, 3.0]
for li in l:
#for i in range(0,10):
#    d = 0.2
#    l = 0.2 + d*i
    x = np.linspace(0.0, li, N)
#    if li == 100.0:
#        plt.plot(x, solExacta(x, A, B, Gamma, li, q), 'k-', lw=4, alpha=0.5, label='L ={:3.3f}'.format(li))
#    else:
    plt.plot(x, solExacta(x, A, B, Gamma, li, q), '-', label='L ={:3.3f}'.format(li))

plt.suptitle('Solución exacta', fontsize = 18, y = 1.0, va = 'top')
plt.title('Difusión estacionaria: $\Gamma = 1.0$, $q = 0.0$', fontsize=12, color='blue')
plt.ylabel('$\psi$')
plt.xlabel('$x$')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(color='white')
plt.savefig('Exacta_SinFuente_Lvariable.pdf', bbox_inches='tight')
```
