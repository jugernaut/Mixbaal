```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['Solarize_Light2'])
```

```python
def solExacta(x, A, B, Gamma, L, t):
    N = 100
    p = np.zeros(len(x))
    PI = np.pi
    for i,xi in enumerate(x):
        xti = 0
        for n in range(1,N):
            xti += np.exp(-n*n*PI*PI*Gamma*t/(L*L)) * \
                   np.sin(n*PI*xi/L) / n
        p[i] = A + (B - A) * (xi / L + 2 * xti / PI)
    return p
```

```python
Gamma = 1.0
L = 1.0
A = 1.0
B = 0.0

N = 100
x = np.linspace(0,L,N)

t = np.array([0.0001, 0.0025,0.0075, 0.0150, 0.0300, 0.100, 0.5])
```

```python
for ti in t:
    plt.plot(x, solExacta(x, A, B, Gamma, L, ti), label='$t$ = {:3.4f}'.format(ti))

plt.suptitle('Solución exacta', fontsize = 18, y = 1.0, va = 'top')
plt.title('Difusión NO estacionaria: $\Gamma =${}, $q = 0.0$'.format(Gamma), fontsize=12, color='blue')
plt.ylabel('$\psi$')
plt.xlabel('$x$')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(color='white')
plt.savefig('Exacta_Dif_NoSta.pdf', bbox_inches='tight')
```
