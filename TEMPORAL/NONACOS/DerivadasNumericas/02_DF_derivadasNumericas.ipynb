# Masa y Densidad.

Suponga que tenemos una barra de longitud $L$. Se puede definir la masa de la barra desde un extremo hasta un punto $x$ en la barra mediante una función $m(x)$, vea la figura siguiente

## Ejercicio 1.

Suponga que $m(x) = -2.5(x-6)^2 + 100$

Escriba una función para calcular $m(x)$ y grafíquela para $x \in [0, 10]$.
Etiquete el eje $x$ y el eje $y$ y ponga un título a la gráfica. 

```python
masa = lambda x: (x+0)**3 + 40*np.fabs(np.cos(x)**2)
```

```python
L = 10
N = 100
x = np.linspace(0, 10, N)

plt.plot(x, masa(x))
plt.xlabel('$x$')
plt.ylabel('$m$')
plt.title('Masa de la barra')
plt.show()
```

## Ejercicio 2.
Dado que se considera la barra en una dimensión, es posible definir una densidad *lineal* como:

$$
\rho(x) = \frac{\text{Masa}}{Longitud} = \frac{m(x)}{x}
$$

Recordemos que la masa es una **propiedad extensiva**: es directamente proporcional al tamaño del sistema o a la cantidad de material, mientras que la densidad es una **propiedad intensiva**: no depende del tamaño del sistema, es invariante a la escala.

Por lo tanto, la densidad se puede calcular en un punto

```python
def densidad(x):
    """ La densidad lineal de la barra.
    Es la derivada de la función de masa.
    """
#    return -5 * (x-6)
    return -5 * (x - 6) + 20 * np.cos(x)
```

```python
def recta(x,x0,x1,y0,y1):
    """
    Dados dos puntos en el plano, cálcula la pendiente de
    la recta que pasa por esos puntos y regresa los puntos
    que pasan por esa recta.
    """
    return ((y1-y0)/(x1-x0))*(x-x0) + y0
```

```python
Lx = 10.0
Nx = 20
x = np.linspace(0,Lx,Nx)
h = Lx / (Nx - 1)
```

```python
m = masa(x)
d = densidad(x)
```

```python
plt.plot(x,m,'.-',lw=2,label='Masa')
plt.plot(x,d,'.-',lw=2,label='Densidad')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('masa')
plt.legend()
plt.show()
```

## Error cuadrático medio (Root Mean Squared Error)
$\displaystyle RMS = \sqrt{\frac{1}{N} \sum_{i=0}^N (ue_i - un_i)^2}$ 

```python
def errorRMS(ue, un):
    """ Error cuadrático medio
    ue : valor exacto
    un : valor numérico aprximado
    """
#    return np.linalg.norm(ue-un) / np.sqrt(len(ue))
    return np.sqrt(((ue - un) ** 2).mean())
```

```python
rmsf = errorRMS(d,df)
rmsb = errorRMS(d,db)
rmsc = errorRMS(d,dc)

print(rmsf,rmsb,rmsc)
```
